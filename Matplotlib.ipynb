## Matplotlib example workflow

```python
#0 import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1 prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. setup plot
fig, ax=plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# customize plot
ax.set(title="Simple Plot",xlabel="x-axis",ylabel="y-axis",)

# save and show
fig.savefig("images/sample-plot.png")
```

## Making figure with numpy array

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
x
```

```python
# plot data
fig, ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# Use the same data to make scatter
fig, ax =plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig,ax=plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
## Make a plot from dictionary
fruits ={"apple" :150,
         "banana": 60,
         "orange":80}

fig, ax= plt.subplots()
ax.bar(fruits.keys(), fruits.values())
ax.set(title="Fruits price" ,ylabel="Price in $ ", xlabel="fruits per Kg");
```

```python
fig, ax= plt.subplots()
ax.barh(list(fruits.keys()), list(fruits.values()))
```

```python
# make some data
x=np.random.randn(1000)
fig,  ax=plt.subplots()
ax.hist(x);
```

```python
 # two options for subplots
fig, ((ax1, ax2), (ax3, ax4) )= plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

# plot to each different axis

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(fruits.keys(), fruits.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot options 2
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
ax[0,0].plot(x,x/2)
```
