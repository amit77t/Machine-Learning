### which one should you used

* when plotting something quickly okay to use pyplot method
* when plotting something advance  okay to use OO method

```python
over_50= heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter", 
             x='age',
             y="chol",
             c='target');
```

```python
# oo method

%matplotlib inline
import matplotlib.pyplot as plt
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
```

```python
fig, ax =plt.subplots(figsize=(10,6))

scatter= ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
ax.set(title="Heart Disease and cholesterol Levels",xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target");

ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
over_50.head()
```

```python
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
ax0.set(title="Heart Disease and cholesterol Levels",xlabel="Age", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target");

ax0.axhline(over_50["chol"].mean(),
           linestyle="--");


scatter= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);

ax1.set(title="Heart Disease and Max Heart Rate",xlabel="Age", ylabel="Max Heart rate")
ax1.legend(*scatter.legend_elements(), title="Target");

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");

fig.suptitle("Heart disease Analysis", fontsize=19, fontweight="bold");
```
